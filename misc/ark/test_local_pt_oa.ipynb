```python
# get the system impact

prompt_system_impact = promtCreator.execute_step("systems_impacted") # promptcreator > executer > goes to dispatcher > analyse function > prompt_Service > (static method) getAnalyseFunction for Prompt Template > checks if req > dump data as json into a Keyword (of langchain PromptTemplate) into prompt template > get Analyse function also returns PromptTemplate.from_template(prompt) > prompt  
text = openai_service.get_4_response(prompt_system_impact)
system_impact_resp = json.loads(text.response)
```

```python
# get the events

prompt_events = promtCreator.execute_step("events")
text = openai_service.get_4_response(prompt_events)
events_resp = json.loads(text.response)
logging.debug("%s %s", "Events: ", events_resp)
```

```python
# get the roles

prompt_roles = promtCreator.execute_step("roles")
text = openai_service.get_4_response(prompt_roles)
roles_resp = json.loads(text.response)
logging.debug("%s %s", "Roles: ", roles_resp)
```

```python
# update the data payload with events data

subset_data['events_data'] = events_resp
promtCreator.set_data(subset_data)
prompt_activities = promtCreator.execute_step("activities")
text = openai_service.get_4_response(prompt_activities)
activities_resp = json.loads(text.response)
logging.debug("%s %s", "Activities: ", activities_resp)
```

```python

# update the data payload for the service catalog

subset_data['events_data'] = events_resp
subset_data['roles_data'] = roles_resp
subset_data['activities_data'] = activities_resp
subset_data['systems_impacted'] = system_impact_resp
promtCreator.set_data(subset_data.copy())  # copy the dict, as we want to re-use for L1 and L2 impacts
```

```python
# get the services 

prompt_services = promtCreator.execute_step("service_catalog")
text = openai_service.get_4_response(prompt_services)
services_resp = json.loads(text.response)
logging.debug("%s %s", "Service Catalog: ", services_resp)
```

```python
# uodate the subset for l1 l2 

subset_data['events_data'] = events_resp
subset_data['roles_data'] = roles_resp
subset_data['activities_data'] = activities_resp
subset_data['systems_impacted'] = system_impact_resp
```

```python

# level 1 impact analysis - reuse subset data

promtCreator.set_data(subset_data.copy())
prompt_l1_impactAnalysis = promtCreator.execute_step("capability_L1_impact_analysis",
                                                             docgen.load_reference_data_files())
text = openai_service.get_4_response(prompt_l1_impactAnalysis)
l1_impacts = json.loads(text.response)
logging.debug("%s %s", "Level 1 Impact Analysis: ", l1_impacts)
```

```python
# filter l1 

l1_impacts_subset = docgen.extract_impacted_capabilities_L1(l1_impacts)
```

```python
# filter l2 

relevant_l2_capabilities = docgen.filter_l2_capabilities_for_l1(l1_impacts_subset)
```

```python
# level 2 impact analysis

promtCreator.set_data(subset_data.copy())
prompt_l2_impactAnalysis = promtCreator.execute_step("capability_L2_impact_analysis", relevant_l2_capabilities)
text = openai_service.get_4_response(prompt_l2_impactAnalysis)
l2_impacts = json.loads(text.response
logging.debug("%s %s", "Level 2 Impact Analysis: ", l2_impacts)
```
