```python

#utils.py

# confluence post proc utils 


 # import os or sys and set os.path() or sys.path as pwd to use this
import json
import requests
from requests.auth import HTTPBasicAuth
from bs4 import BeautifulSoup



def fetch_page_data(secrets, page_id):

    PAGE_ID_OR_TITLE = page_id


    CONFLUENCE_EMAIL = secrets["CONFLUENCE_EMAIL"]
    CONFLUENCE_TOKEN = secrets["CONFLUENCE_TOKEN"]
    BASE_URL = secrets["BASE_URL"]
    SPACE_KEY = secrets["SPACE_KEY"]
    
    BASIC_AUTH = HTTPBasicAuth(CONFLUENCE_EMAIL, CONFLUENCE_TOKEN)
    HEADERS = {'Content-Type': 'application/json;charset=utf-8'}
    
    API_URL = f"{BASE_URL}/rest/api/content/{PAGE_ID_OR_TITLE}?expand=body.view,version"
    response = requests.get(
            API_URL,
            headers=HEADERS,
            auth=BASIC_AUTH
        )
    if response.status_code == 200:
        page_data = response.json()
        print("All Page Details returned in json.")
        # print_confluence_details(page_data)
        return page_data
    else:
        print(f"Error: {response.status_code} - {response.text}")
        return None

# following codes used to extract tables insides sections and sections into dictionary format

def extract_table(table):
    headers = [th.get_text().strip() for th in table.find_all('th')]
    rows = []
    for tr in table.find_all('tr')[1:]:  # Skip header row
        cells = tr.find_all('td')
        if len(cells) == len(headers):
            row_data = {headers[i]: cells[i].get_text().strip() for i in range(len(cells))}
            rows.append(row_data)
    table_dict = {"column_headers": headers, "rows": rows}
    return table_dict
   

def extract_sections(soup):
    sections = {}
    current_section = None
    current_content = []

    for element in soup.find_all(['h1', 'h2', 'h3', 'h4', 'h5', 'h6', 'p', 'table', 'div']):
        if element.name in ['h1', 'h2', 'h3', 'h4', 'h5', 'h6']:
            if current_section:
                sections[current_section] = current_content
            current_section = element.get_text().strip()
            current_content = []
        elif current_section:
            if element.name == 'p':
                current_content.append(element.get_text().strip())
            elif element.name == 'table':
                table_data = extract_table(element)
                current_content.append(table_data)
            elif element.name == 'div' and 'table-wrap' in element.get('class', []):
                table_data = extract_table(element.find('table'))
                current_content.append(table_data)

    if current_section:
        sections[current_section] = current_content

    # Process contents
    processed_sections = {}
    for title, content in sections.items():
        if content:
            processed_content = []
            for item in content:
                if isinstance(item, dict):  # It's a table # then convert to string as its giving an issue while embedding later
                    processed_content.append(str(item))
                else:  # It's a string
                    processed_content.append(item)
            processed_sections[title] = processed_content
    return processed_sections


def create_vdb_data(shortlisted_page_ids, aggregated_pages={}):
    '''
    input : shortlisted_page_ids
    aggregated_pages : output > can either be input of empty dictionary or existing dictioanry
    '''
    for page_id in shortlisted_page_ids:
        
        space_key = secrets["SPACE_KEY"]        
        page_dict = {}
        PAGE_ID_OR_TITLE = page_id  # Replace with the actual page ID or title
        
        page_data = fetch_page_data(secrets, page_id)
        
        page_url = f"{BASE_URL}/spaces/{space_key}/pages/{page_id}"
        
        page_title=page_data['title']
        
        page_author=page_data.get("version", {}).get("by", {}).get("publicName", None)
        
        content_value = page_data.get('body', {}).get('view', {}).get('value', '')
        
        page_dict['title'] = page_title
        page_dict['author'] = page_author
        page_dict['html_content'] = content_value
        page_dict['page_id'] = page_id 
        page_dict['page_url'] = page_url
        
        soup = BeautifulSoup(content_value, 'html.parser')
    
        sections_dict = extract_sections(soup)    # this coe scrapes all possible headings to make a dictionary of headings as keys and its contents as value 
        
        page_dict["all_sections"] = sections_dict # this dict has Problem, Context, Solutions and key reason/s, Options as subheadings

        # all if statements to ensure that we skip over and store all_sections, for all ADRs , whether or not they have any of these sections well formatted 
        
        if "Problem" in sections_dict.keys():
            page_dict["problem"] = { "content" : sections_dict["Problem"],
                                 "tag" : "problem_statement",
                                 "page_ref" : page_id , # for fetching entire page later
                                 "ad_ref" : page_title, 
                                 "page_link" : page_url }
    
        if "Context" in sections_dict.keys():
            page_dict["context"] = { "content" : sections_dict["Context"],
                                 "tag" : "context",
                                 "page_ref" : page_id , # for fetching entire page later
                                 "ad_ref" : page_title, 
                                 "page_link" : page_url }
    
        if "Solution & key reason/s" in sections_dict.keys() : 
            page_dict["solution_&_key_reasons"] =  { "content" : sections_dict["Solution & key reason/s"],
                                 "tag" : "solution",
                                 "page_ref" : page_id , # for fetching entire page later
                                 "ad_ref" : page_title, 
                                 "page_link" : page_url }
    
        if "Solution" in sections_dict.keys() : 
            page_dict["solution_&_key_reasons"] =  { "content" : sections_dict["Solution"],
                                 "tag" : "solution",
                                 "page_ref" : page_id , # for fetching entire page later
                                 "ad_ref" : page_title, 
                                 "page_link" : page_url }
    
        if "Options" in sections_dict.keys():
            page_dict["options"] = { "content" : sections_dict["Options"],
                                 "tag" : "options",
                                 "page_ref" : page_id , # for fetching entire page later
                                 "ad_ref" : page_title, 
                                 "page_link" : page_url }
        
        
        aggregated_pages.update({page_id : page_dict})

    return aggregated_pages
        
            
```

```python

def extract_pcos_content_from_all_sections_2(sections_dict, page_id, page_title, page_url):
    '''
    takes in all_Sections dict : output of  extract_sections(soup)  and converts into the content wise page_dict 
    '''
    # Define keywords for different sections
    section_keywords = {
        "problem": ["problem"],
        "context": ["context"],
        "options": ["options"],
        "solution": ["solution"]
    }

    # Initialize dictionaries to store content for each section
    page_pcos_dict = {keyword: {"content": "", "tag": keyword, "page_ref": page_id, "ad_ref": page_title, "page_link": page_url} for keyword in section_keywords}

    # trying to solve for list of strings 
    def fetch_strings_from_list(input_list):
        return ' '.join(item for sublist in input_list for item in (sublist if isinstance(sublist, list) else [sublist]))
    
    # Iterate through the keys in sections_dict
    for key in sections_dict.keys():
        # Check if the key contains any of the section keywords
        for section, keywords in section_keywords.items():
            if any(keyword in key.lower() for keyword in keywords):
                # Concatenate the value associated with this key to the corresponding section
                page_pcos_dict[section]["content"] += key + " : " + fetch_strings_from_list(sections_dict[key]) + " " 
                
                

    # Remove trailing newline characters from content
    for section in page_pcos_dict:
        page_pcos_dict[section]["content"] = page_pcos_dict[section]["content"].strip()

    return page_pcos_dict


def create_vdb_data_2(shortlisted_page_ids, aggregated_pages={}):
    '''
    input : shortlisted_page_ids
    aggregated_pages : output > can either be input of empty dictionary or existing dictioanry
    '''
    for page_id in shortlisted_page_ids:
        
        space_key = secrets["SPACE_KEY"]        
        page_dict = {}
        PAGE_ID_OR_TITLE = page_id  # Replace with the actual page ID or title
        
        page_data = fetch_page_data(secrets, page_id)
        
        page_url = f"{BASE_URL}/spaces/{space_key}/pages/{page_id}"
        
        page_title=page_data['title']
        
        page_author=page_data.get("version", {}).get("by", {}).get("publicName", None)
        
        content_value = page_data.get('body', {}).get('view', {}).get('value', '')
        
        page_dict['title'] = page_title
        page_dict['author'] = page_author
        page_dict['html_content'] = content_value
        page_dict['page_id'] = page_id 
        page_dict['page_url'] = page_url
        
        soup = BeautifulSoup(content_value, 'html.parser')
    
        sections_dict = extract_sections(soup)    # this code scrapes all possible headings to make a dictionary of headings as keys and its contents as value 
        
        page_dict["all_sections"] = sections_dict # this dict has Problem, Context, Solutions and key reason/s, Options as subheadings

        # all if statements to ensure that we skip over and store all_sections, for all ADRs , whether or not they have any of these sections well formatted 
        
        page_dict_content = extract_pcos_content_from_all_sections_2(sections_dict, page_id, page_title, page_url)

        page_dict.update(page_dict_content)
        
        aggregated_pages.update({page_id : page_dict})

    return aggregated_pages
        
```

```python
#VECTOR_DB UTILS.PY 

from qdrant_client import QdrantClient
from qdrant_client.http import models 
from qdrant_client.http.models import CollectionStatus

from sentence_transformers import SentenceTransformer

import hashlib

def get_qdrant_client(host, port):
    client = QdrantClient(host="localhost", port = 6333)
    return client

# so that same page_id doesnt rewrite all content
def get_consistent_id(page_id, subheading):
    unified_str = str(page_id) + str(subheading)
    hash_id = int(hashlib.md5(unified_str.encode()).hexdigest(), 16) % (10**8)
    return hash_id


def create_adr_collection(client, collection_name, vector_size):
    try:
        if not client.collection_exists(collection_name= collection_name):
            client.create_collection(
            collection_name=f"{collection_name}",
            vectors_config=models.VectorParams(
                                        size=vector_size,                        # depending on embedding model
                                        distance=models.Distance.COSINE,
                                        # datatype=models.Datatype.UINT8, # if vectors are int8 quantised , eg. cohere
                                    ),
            sparse_vectors_config={
                "text": models.SparseVectorParams(),
                },
            )
    
        print(f"Collection '{collection_name}' created successfully.")
    except Exception as e:
        print(f"Collection '{collection_name}' already exists or an error occurred: {e}")



# add each page to vdb
def add_content_to_qdrant(client, collection_name, hash_id, page_data, subheading_data, embedding_model):
    content = subheading_data["content"]    
    metadata = {
        "tag" : subheading_data["tag"],
        "page_ref" : subheading_data["page_ref"],
        "ad_ref" : subheading_data["ad_ref"],       # title is the AD reference
        "page_link" : subheading_data["page_link"],
        "author" : page_data["author"]
        
    }

    #embeddings
    embedding_list = embedding_model.encode(content).tolist()[0] # embedding list must be a [non nested list]. first element[0] must be a float
    # handle blank content embedding that returns a single float number 
    # if isinstance(embedding_list, float):
    #     embedding_list = [embedding_list]

    if isinstance(embedding_list, list): # upsert only if there is a valid embedding created for that subheading
        #insertion
        client.upsert(
            collection_name = collection_name, 
            points = [{
                "id" : hash_id,
                "vector" : embedding_list,
                "payload" : metadata
            }]
        )


def vdb_aggregate_ingestion(agg_pages, collection_name, embedding_model):
    '''
    agg_pages : output of the create_vdb_data function 
    no output : adds data to local instance of qvd
    '''
    #adding all pages, subheadings, subhead metadata to vdb  
    for page_id, page_data in agg_pages.items(): # dictionary output of create_vdb function
        for subheading, subheading_data in page_data.items():
            if isinstance(subheading_data, dict) and "content" in subheading_data:
                hash_id = get_consistent_id(page_id, subheading_data["tag"]) 
                add_content_to_qdrant(client, collection_name, hash_id, page_data, subheading_data, embedding_model)
                # print(page_id, type(page_id),'\n', subheading_data, '\n')\
    print("Content added to Qdrant.")

```
