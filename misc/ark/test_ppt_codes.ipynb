### Extract All Metadata from a Powerpoint file 

```python

def extract_table_data(table):
    """
    Extracts cell-wise data from a table and returns it as a dictionary.
    """
    table_dict = {}
    cell_id = 1

    for row in table.rows:
        for cell in row.cells:
            table_dict[f'cell_{cell_id}'] = cell.text.strip()
            cell_id += 1
    
    return table_dict

def extract_shape_data(shape):
    """
    Extracts data from a shape and returns it in a dictionary with details.
    """
    shape_data = {}
    
    if shape.has_text_frame and shape.text_frame.text.strip():
        shape_data['type'] = 'TEXT_BOX'
        shape_data['content'] = shape.text_frame.text.strip()
    
    elif shape.has_table:
        shape_data['type'] = 'TABLE'
        shape_data['content'] = extract_table_data(shape.table)
    
    elif shape.shape_type == 1:  # AutoShape
        if shape.text_frame and shape.text_frame.text.strip():
            shape_data['type'] = 'AUTO_SHAPE'
            shape_data['content'] = shape.text_frame.text.strip()
    
    elif shape.shape_type == 6:  # Group Shape
        group_shapes = {}
        for i, sub_shape in enumerate(shape.shapes, 1):
            group_shapes[f'SubShape_{i}'] = extract_shape_data(sub_shape)
        shape_data['type'] = 'GROUP'
        shape_data['content'] = group_shapes
    
    return shape_data

def extract_slide_shapes(slide):
    """
    Extracts all shapes from a slide and returns a dictionary with each shape's data.
    """
    slide_data = {}
    
    for idx, shape in enumerate(slide.shapes, 1):
        slide_data[f'Shape_{idx}'] = extract_shape_data(shape)
    
    return slide_data
```
