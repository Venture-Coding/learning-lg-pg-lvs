```python
def extract_sections_test(sections_dict, page_id, page_title, page_url):
    # Define keywords for different sections
    section_keywords = {
        "problem": ["problem"],
        "context": ["context"],
        "options": ["options"],
        "solution": ["solution"]
    }

    # Initialize dictionaries to store content for each section
    page_dict = {keyword: {} for keyword in section_keywords}

    # Iterate through the keys in sections_dict
    for key in sections_dict.keys():
        # Check if the key contains any of the section keywords
        for section, keywords in section_keywords.items():
            if any(keyword in key.lower() for keyword in keywords):
                # Append the value associated with this key to the corresponding section
                page_dict[section]["content"] = sections_dict[key]
                page_dict[section]["tag"] = f"{section}"
                page_dict[section]["page_ref"] = page_id
                page_dict[section]["ad_ref"] = page_title
                page_dict[section]["page_link"] = page_url

    return page_dict

# Example usage
sections_dict = {
    "Introduction": "This is an introduction.",
    "Problem Statement": "This is the problem statement.",
    "Context": "This is the context section.",
    "Options": "These are the available options.",
    "Solution": "Here's the proposed solution.",
    "Problem" : "problem statement 2.", 
    "problems" : "problem statement 3.", 
    "the options" : "option statement 2.",
    "solution and key ": " solution statement 2." 
}

page_id = "123"
page_title = "Sample Page"
page_url = "1"

result = extract_sections_test(sections_dict, page_id, page_title, page_url)
print(result)
```
